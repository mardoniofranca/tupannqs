```python
import pandas as pd
import os
import numpy as np
```

```python
source = "../../../../data/dfs/"
dest   = "../../../../data/total/df_0304.csv"
```

```python
def append(source, dest):
    arr = os.listdir(source)
    df = pd.DataFrame()
    for a in arr:
        try:        
            path = source + a
            df_aux = pd.read_csv(path)
            df = pd.concat([df, df_aux], ignore_index=True, sort=False) 
        except :
            print(a)
    return df
```

```python
df = append(source,dest)
```
